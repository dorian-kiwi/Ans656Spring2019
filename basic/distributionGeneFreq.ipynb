```julia
using Distributions, Plots
```

```julia
N  = 100    # population size
p0 = zeros(N+1)
p0[Int(N/2)+1] = 1.0 
pDist = copy(p0)
p = collect(0:N)/N
res = [pdf.(Binomial(N,prob),0:N) for prob in p]
M   = reduce(hcat, res);
```

```julia
plot(pDist)
```

```julia
pDist = M*pDist
plot(pDist)
```

```julia
pDist = M*pDist
plot(pDist)
```

```julia
n = 2
plot(M^n * p0)
```
