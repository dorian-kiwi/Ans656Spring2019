### Function to read in genotype matrix

```julia
function readMatBin(fileName)
    genStr = open(fileName)
    n = read(genStr,Int64)
    p = read(genStr,Int64)
    M = zeros(n,p)
    for j in 1:p
        for i in 1:n
            M[i,j] = read(genStr,Float64)
        end
    end
    close(genStr)
    return M
end
```

### Reading in data files

```julia
y       = readdlm("phenotypes.csv")
Markers = readMatBin("genotypes.bin")
M = Markers .- mean(Markers,dims=1)
mean(M,dims=1)
```

## Initialize variables

```julia
nIter  = 2000
burnin = 101
n,k = size(M)
mean2pq = mean(var(M,dims=1))
Ve  = var(y)/2
Va  = Ve/(k*mean2pq)       
α   = zeros(k)
mu  = 0
meanAlpha = zeros(k)
meanMu = 0
w = y .- mu - M*α; # y adjusted for intercept and marker effects
samplesVa = zeros(nIter)
samplesVe = zeros(nIter)
samplesα  = zeros(k,nIter)
νa = 4.0
νe = 4.0
S2a = Va*(νa-2)/νa
S2e = Ve*(νe-2)/νe
```

## Loop for MCMC samples

```julia
@showprogress "obtaining MCMC samples for BayesC0..." for i=1:nIter
    # sample intercept
    global w,mu, meanMu, α, meanAlpha, Ve, Va
    w = w .+ mu                   # add the effect of mu 
    meanMu = mean(w)
    std = sqrt(Ve/n)
    mu = meanMu + randn()*std
    meanMu += (mu - meanMu)/i
    w = w .- mu                  # adjust for sampled mu
    # loop for sampling marker effects
    λ = Ve/Va
    for j = 1:k
        x = M[:,j]
        w = w .+ x*α[j]          # add the effect of marker effect j
        lhs = (x'x + λ)[1]
        rhs = (x'w)[1]
        meanAlpha_j = rhs/lhs
        std = sqrt(Ve/lhs)
        α[j] = meanAlpha_j + randn()*std
        w -= x*α[j]              # adjust for marker effect j
    end
    samplesα[:,i] = α
    # sample Va
    Va = (α'α + νa*S2a)/rand(Chisq(k+νa),1)[1]
    samplesVa[i] = Va
    # sample Ve
    Ve = ((w'w)[1] + νe*S2e)/rand(Chisq(n+νe),1)[1]
    samplesVe[i] = Ve
end
```

```julia
histogram(samplesVa,nbins=50,normed=true,
title  = "Posterior Distributions of Va",
ylabel = "Frequency", 
xlabel = "Va")
```

```julia
histogram(samplesVe,nbins=50,normed=true,
title  = "Posterior Distributions of Ve",
ylabel = "Frequency", 
xlabel = "Ve")
```

### Function for calculating proportion of variance for genomic window  

```julia
function winPPA(M,samplesα,win)
    wPPA = zeros(nIter)
    nSamples = size(samplesα,2)
    for i=1:nSamples
        genVar = var(M*samplesα[:,i])
        winVar = var(M[:,win]*samplesα[win,i])
        wPPA[i] = winVar/genVar
    end
    wPPA
end
```

### Calculate posterior probability that window explains more than 10% of total variance

```julia
wStarts = [80, 100, 120,300,400,500]            # starting positions of windows
wSize   = 20                                    # window size
[sum(winPPA(M,samplesα,i:i+wSize)[burnin:end] .> 0.1)/(nIter-burnin+1) for i in wStarts]
```
